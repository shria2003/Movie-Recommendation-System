```python
content_based_dataset=d4.drop(columns=['belongs_to_collection', 'budget', 'homepage',
       'original_language','original_title','popularity', 'poster_path', 'production_companies',
       'production_countries', 'revenue', 'runtime',
       'spoken_languages', 'status', 'tagline', 'video',
       'vote_average', 'vote_count'])

content_based_dataset.dropna(inplace=True)
content_based_dataset
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd


# Create DataFrame
df = pd.DataFrame(content_based_dataset)

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the overview data
tfidf_matrix = tfidf_vectorizer.fit_transform(df['overview'])

# Compute cosine similarity between all pairs of movies
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_sim
```

```python
genres = [{'id': 16, 'name': 'Animation'}, {'id': 35, 'name': 'Comedy'}, {'id': 10751, 'name': 'Family'}]

names = [genre['name'] for genre in genres]

print(names)
```

```python

# Function to recommend similar movies based on overview
def recommend_similar_movies(movie_index, similarity_matrix, df):
    # Get similarity scores for the given movie index
    sim_scores = list(enumerate(similarity_matrix[movie_index]))
    # Sort movies based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get top 5 similar movies (excluding the movie itself)
    # sim_scores = sim_scores[1:6]
    # Get movie indices
    movie_indices = [i[0] for i in sim_scores]
    # Get movie titles
    similar_movies = df[['id','title']].iloc[movie_indices]
    return similar_movies

# Example: Recommend similar movies to the movie at index 0
similar_movies = recommend_similar_movies(1, cosine_sim, df)
print("Recommended similar movies:")
print(similar_movies)
```

## Collaborative Filtering

```python
ratings_data = pd.read_csv('ratings.csv')
ratings_data.rename(columns={'movieId': 'id'}, inplace=True)
```

```python
# Calculate the popularity of each movie based on the number of ratings
movie_popularity = ratings_data.groupby('id').size().reset_index(name='popularity')
movie_popularity
```

```python
# Merge popularity with ratings data
ratings_data = pd.merge(ratings_data, movie_popularity, on='id')
ratings_data
```

```python

# Calculate the mean rating for each movie
movie_ratings_mean = ratings_data.groupby('id')['rating'].mean().reset_index(name='rating_mean')

# Merge mean ratings with ratings data
ratings_data = pd.merge(ratings_data, movie_ratings_mean, on='id')
ratings_data
```

```python
# Filter out movies with low popularity (you can adjust this threshold)
popular_movies = ratings_data[ratings_data['popularity'] >= 50]
# popular_movies.drop(columns=['rating_mean_x','rating_mean_y'], inplace=True)
```

```python
# Collaborative Filtering based on popularity and rating
def collaborative_filtering(ratings_data, user_id, n_recommendations=10):
    # Filter out movies rated by the user
    user_ratings = ratings_data[ratings_data['userId'] == user_id]
    
    # Sort by popularity and rating mean
    user_ratings_sorted = user_ratings.sort_values(by=['popularity', 'rating_mean'], ascending=False)
    
    # Get top N rated movies
    top_n_movies = user_ratings_sorted.head(n_recommendations)
    
    return top_n_movies

# Example: Recommend movies for user 1
user_id = 1
recommendations = collaborative_filtering(popular_movies, user_id)
print("Recommendations for User", user_id, ":")
print(recommendations)
```

## Hybrid Filtering

```python
# Hybrid Filtering
def hybrid_filtering(collaborative_recommendations, content_based_recommendations):
    # Merge recommendations from both approaches
    hybrid_recommendations = pd.merge(collaborative_recommendations, content_based_recommendations, on='id')
    
    return hybrid_recommendations

# Example: Hybrid Recommendations for User 1
user_id = 1

# Collaborative Filtering recommendations
collaborative_recommendations = collaborative_filtering(popular_movies, user_id)

# Content-Based Filtering recommendations
# We'll use the function recommend_similar_movies from the content-based approach
# Let's assume we want recommendations based on the movie at index 6 (or any other movie)
similar_movies = recommend_similar_movies(user_id, cosine_sim, df)

# Merge recommendations from both approaches using the movieId
content_based_recommendations = pd.DataFrame({'id': similar_movies.index})
content_based_recommendations['id'] = content_based_recommendations['id'] + 1  # Adjust movieId to match ratings data

# Hybrid Filtering
hybrid_recommendations = hybrid_filtering(collaborative_recommendations, content_based_recommendations)

print("Hybrid Recommendations for User", user_id, ":")
print(hybrid_recommendations)
```
